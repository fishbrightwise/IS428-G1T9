```python
itm_df = pd.DataFrame(pd.read_csv('../data/imdb_tmdb_cleaned_w|o_filtering.csv'))
print(itm_df.shape)
itm_df.head()
```

```python
rtm_df = pd.DataFrame(pd.read_csv('../data/rotten_tomatoes_movies_cleaned.csv'))
print(rtm_df.shape)
rtm_df.head()
```

```python
itm_df.columns
```

### Attribute Processing

```python
itm_df.head()
```

```python
itm_df['production_companies'] = itm_df['production_companies'].progress_apply(ast.literal_eval)
itm_df['production_countries'] = itm_df['production_countries'].progress_apply(ast.literal_eval)
itm_df['actor_list'] = itm_df['actor_list'].progress_apply(ast.literal_eval)
itm_df['keywords'] = itm_df['keywords'].progress_apply(ast.literal_eval)
itm_df['genre'] = itm_df['genre'].progress_apply(ast.literal_eval)
itm_df['director'] = itm_df['director'].progress_apply(ast.literal_eval)
itm_df.head()
```

```python
rtm_df.head()
```

```python
rtm_df.columns
```

```python
rtm_df = rtm_df[['rotten_tomatoes_link', 'movie_title', 'tomatometer_status', 'tomatometer_rating', 'audience_status', 'audience_rating', 'compound_score', 'sentiment']]
rtm_df.head()
```

```python
merged_df = pd.merge(itm_df, rtm_df, left_on='title', right_on='movie_title', how='inner')
merged_df = merged_df.drop(columns=['movie_title'])
print(merged_df.shape)
```

### Exporting Data

```python
itm_df.to_csv('../data/imdb_tmdb_regression.csv', index=False)
```

```python
merged_df.to_csv('../data/merged_regression.csv', index=False)
```
